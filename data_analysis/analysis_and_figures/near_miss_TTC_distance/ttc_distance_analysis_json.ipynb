## Load Processed Data of NDE and D2RL

```python
import numpy as np
root_folder = "../../processed_data/"

NDE_near_miss_ttc = np.load(root_folder + "NDE/NDE_ttc_near_miss.npy")
NDE_near_miss_distance = np.load(root_folder + "NDE/NDE_distance_near_miss.npy")
distance_list_all_experiments = np.load(root_folder + "D2RL/NADE_distance_near_miss.npy")
ttc_list_all_experiments = np.load(root_folder + "D2RL/NADE_ttc_near_miss.npy")
distance_weight_list_all_experiments = np.load(root_folder + "D2RL/NADE_distance_near_miss_weight.npy")
ttc_weight_list_all_experiments = np.load(root_folder + "D2RL/NADE_ttc_near_miss_weight.npy")
```

## TTC D2RL Weighted Distribution and NDE Distribution

```python
font = {'family' : 'Times New Roman',
        'size'   : 15}
color_blue = '#130074' # deep blue|
color_red = (131/255,5/255,24/255)

import matplotlib.pyplot as plt
plt.rc('font', **font)

from mpl_toolkits.axes_grid1 import Divider, Size
fig = plt.figure(figsize=(8,8))
color_red = (131/255,5/255,24/255)
# The first items are for padding and the second items are for the axes.
# sizes are in inch.
h = [Size.Fixed(1.0), Size.Fixed(5.5)]
v = [Size.Fixed(1.0), Size.Fixed(4)]
divider = Divider(fig, (0, 0, 1, 1), h, v, aspect=False)
# The width and height of the rectangle are ignored.
ax1 = fig.add_axes(divider.get_position(), axes_locator=divider.new_locator(nx=1, ny=1))

bins = np.linspace(0, 5, 50)
ax1.hist(list(NDE_near_miss_ttc), density=True, bins=bins, color=color_blue, alpha=0.75, label="NDE")
ax1.hist(ttc_list_all_experiments, weights=ttc_weight_list_all_experiments, color=color_red,  density=True, bins=bins, alpha=0.75, label="D2RL")
# plt.yscale("log")
label_flag = True
if label_flag:
        plt.xlabel("Time to collision (s)", fontsize=25)
        plt.ylabel("Frequency", fontsize=25)
plt.legend()
plt.ylim(0, 1.2)
plt.savefig("TTC_weighted_near_miss.svg", bbox_inches = 'tight')
plt.savefig("TTC_weighted_near_miss.pdf", bbox_inches = 'tight')
plt.savefig("TTC_weighted_near_miss.png", bbox_inches = 'tight', dpi=400)
# plt.xlim(0,10)
```
