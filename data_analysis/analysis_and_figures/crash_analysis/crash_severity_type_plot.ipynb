# 1. Crash Severity Analysis

## D2RL Weighted Crash Severity Distribution and NDE Crash Severity Distribution

```python
import numpy
font = {'family' : 'Times New Roman',
        'size'   : 15}

import json
import matplotlib.pyplot as plt
plt.rc('font', **font)

from mpl_toolkits.axes_grid1 import Divider, Size
fig = plt.figure(figsize=(8,8))
color_blue = '#130074' # deep blue|
color_red = (131/255,5/255,24/255)
# The first items are for padding and the second items are for the axes.
# sizes are in inch.
h = [Size.Fixed(1.0), Size.Fixed(5.5)]
v = [Size.Fixed(1.0), Size.Fixed(4)]
divider = Divider(fig, (0, 0, 1, 1), h, v, aspect=False)
# The width and height of the rectangle are ignored.
ax1 = fig.add_axes(divider.get_position(), axes_locator=divider.new_locator(nx=1, ny=1))

import matplotlib.pyplot as plt

# Loading Data
model_based_speed_diff_json_path = "../../processed_data/NDE/crash_type_severity/speed_diff_norm/speed_diff_norm.json"
D2RL_based_speed_diff_json_path = "../../processed_data/D2RL/crash_type_severity/speed_diff_norm/speed_diff_norm.json"

with open(model_based_speed_diff_json_path, "r") as json_file:
    model_json_obj = json.load(json_file)
    model_speed_diff_list = model_json_obj["speed_diff_list"]
    model_weight_list = model_json_obj["weight"]

with open(D2RL_based_speed_diff_json_path, "r") as json_file:
    D2RL_json_obj = json.load(json_file)
    D2RL_speed_diff_list = D2RL_json_obj["speed_diff_list"]
    D2RL_weight_list = D2RL_json_obj["weight"]

# Plot histgram
    
bins = numpy.linspace(0, 20, 15)

w1 = ax1.hist(model_speed_diff_list, bins, alpha=0.75, color=color_blue, label='NDE', density=True)
w2 = ax1.hist(D2RL_speed_diff_list, bins, weights=D2RL_weight_list, color=color_red, alpha=0.75, label='D2RL', density=True)
label_flag=True
# plt.ylim((0,0.35))
if label_flag:
        plt.xlabel("Speed difference (mph)", fontsize=25)
        plt.ylabel("Frequency", fontsize=25)
plt.legend()
# plt.title("Crash Severity Distribution (Raw)")
plt.savefig("severity_weighted_hist.svg", bbox_inches = 'tight')
plt.savefig("severity_weighted_hist.pdf", bbox_inches = 'tight')
plt.savefig("severity_weighted_hist.png", bbox_inches = 'tight', dpi=400)
```

## D2RL Weighted Crash Severity Distribution and NDE Crash Severity Distribution (4 Levels)

# 2. Crash Type Analysis

## D2RL and NDE Weighted Crash Type Distribution

```python
import matplotlib.pyplot as plt
plt.rc('font', **font)
import numpy as np

from mpl_toolkits.axes_grid1 import Divider, Size
fig = plt.figure(figsize=(8,8))
color_red = (131/255,5/255,24/255)
# The first items are for padding and the second items are for the axes.
# sizes are in inch.
h = [Size.Fixed(1.0), Size.Fixed(5.5)]
v = [Size.Fixed(1.0), Size.Fixed(4)]
divider = Divider(fig, (0, 0, 1, 1), h, v, aspect=False)
# The width and height of the rectangle are ignored.
ax = fig.add_axes(divider.get_position(), axes_locator=divider.new_locator(nx=1, ny=1))
fontsize = 15
x1 = np.array([0 + i for i in range(4)])
bar_width = 0.15
# Loading Data
model_based_speed_diff_json_path = "../../processed_data/NDE/crash_type_severity/crash_type/crash_type_res.json"
D2RL_based_speed_diff_json_path = "../../processed_data/D2RL/crash_type_severity/crash_type/crash_type_res.json"
collision_name_list = ["sideswipe", "rear_end", "angle", "head_on"]

with open(model_based_speed_diff_json_path, "r") as json_file:
    model_json_obj = json.load(json_file)

with open(D2RL_based_speed_diff_json_path, "r") as json_file:
    D2RL_json_obj = json.load(json_file)

# Plot figures
result1 = [model_json_obj[name]["weighted_num"] for name in collision_name_list]
result2 = [D2RL_json_obj[name]["weighted_num"] for name in collision_name_list]

result1 = np.array(result1) * 1.6e-7 / sum(np.array(result1))
result2 = np.array(result2) * 1.6e-7 / sum(np.array(result2))

# ax.set_xticks([0 + 0.5 * i for i in range(len(plot_collision_type))])
ax.set_xticks([0, 1, 2, 3])
ax.bar(x1+0.5*bar_width, result1, width=bar_width, color=color_blue, align='center', label="NDE")
ax.bar(x1-0.5*bar_width, result2, width=bar_width, color=color_red, align='center', label="D2RL")
ax.set_xticklabels(["Sideswipe", "Rear End", "Angle", "Head On"])
plt.xlabel("Crash type", fontsize=25)
plt.ylabel("Crash rate", fontsize=25)
ax.legend()
plt.savefig("crash_type_weighted.svg", bbox_inches = 'tight')
plt.savefig("crash_type_weighted.pdf", bbox_inches = 'tight')
plt.savefig("crash_type_weighted.png", bbox_inches = 'tight', dpi=400)
```

## D2RL Raw Crash Severity Distribution and NDE Crash Severity Distribution

```python
import matplotlib.pyplot as plt
plt.rc('font', **font)
import numpy as np

from mpl_toolkits.axes_grid1 import Divider, Size
fig = plt.figure(figsize=(8,8))
color_red = (131/255,5/255,24/255)
# The first items are for padding and the second items are for the axes.
# sizes are in inch.
h = [Size.Fixed(1.0), Size.Fixed(5.5)]
v = [Size.Fixed(1.0), Size.Fixed(4)]
divider = Divider(fig, (0, 0, 1, 1), h, v, aspect=False)
# The width and height of the rectangle are ignored.
ax = fig.add_axes(divider.get_position(), axes_locator=divider.new_locator(nx=1, ny=1))
fontsize = 15
x1 = np.array([0 + i for i in range(4)])
bar_width = 0.15
model_based_speed_diff_json_path = "../../processed_data/NDE/crash_type_severity/crash_type/crash_type_res.json"
D2RL_based_speed_diff_json_path = "../../processed_data/D2RL/crash_type_severity/crash_type/crash_type_res.json"
D2RL_episode_number = 3141500 # The total number of D2RL episodes should be specified here
collision_name_list = ["sideswipe", "rear_end", "angle", "head_on"]

with open(model_based_speed_diff_json_path, "r") as json_file:
    model_json_obj = json.load(json_file)

with open(D2RL_based_speed_diff_json_path, "r") as json_file:
    D2RL_json_obj = json.load(json_file)

result1 = np.array([model_json_obj[name]["num"] for name in collision_name_list])
result2 = np.array([D2RL_json_obj[name]["num"] for name in collision_name_list])

result2 = result2/D2RL_episode_number # The total number of D2RL episodes
print(sum(result2))
# ax.set_xticks([0 + 0.5 * i for i in range(len(plot_collision_type))])
ax.set_xticks([0, 1, 2, 3])
ax.bar(x1, result2, width=bar_width, align='center', color=color_red, label="D2RL")
ax.set_xticklabels(["Sideswipe", "Rear End", "Angle", "Head On"])
plt.xlabel("Crash type", fontsize=25)
plt.ylabel("Crash rate", fontsize=25)
ax.legend()
plt.savefig("crash_type_raw.svg", bbox_inches = 'tight')
plt.savefig("crash_type_raw.pdf", bbox_inches = 'tight')
plt.savefig("crash_type_raw.png", bbox_inches = 'tight', dpi=400)
```
