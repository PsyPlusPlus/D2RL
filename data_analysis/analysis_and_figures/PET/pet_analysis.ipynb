## NDE and D2RL Processed Data Load

```python
import numpy as np
root_folder = "../../processed_data/"
NDE_near_miss_pet = np.load(root_folder + "NDE/NDE_near_miss_pet.npy")

pet_list_all_experiments = np.load(root_folder + "D2RL/NADE_near_miss_pet.npy")
pet_weight_list_all_experiments = np.load(root_folder + "D2RL/NADE_near_miss_pet_weight.npy")
```

## NDE and D2RL Weighted PET Distribution

```python
font = {'family' : 'Times New Roman',
        'size'   : 15}
color_blue = '#130074' # deep blue|
color_red = (131/255,5/255,24/255)

import matplotlib.pyplot as plt
plt.rc('font', **font)

from mpl_toolkits.axes_grid1 import Divider, Size
fig = plt.figure(figsize=(8,8))
color_red = (131/255,5/255,24/255)

# The first items are for padding and the second items are for the axes.
# sizes are in inch.
h = [Size.Fixed(1.0), Size.Fixed(5.5)]
v = [Size.Fixed(1.0), Size.Fixed(4)]
divider = Divider(fig, (0, 0, 1, 1), h, v, aspect=False)
# The width and height of the rectangle are ignored.
ax1 = fig.add_axes(divider.get_position(), axes_locator=divider.new_locator(nx=1, ny=1))
bins = np.linspace(0, 5, 50)
ax1.hist(list(NDE_near_miss_pet), density=True, bins=bins, alpha=0.75, label="NDE", color=color_blue)
ax1.hist(pet_list_all_experiments, weights=pet_weight_list_all_experiments, density=True, bins=bins, alpha=0.75, label="D2RL", color=color_red)
label_flag = True


plt.xlabel("Post encroachment time (s)", fontsize=25)
plt.ylabel("Frequency", fontsize=25)
plt.legend()
plt.ylim((0,1.2))
plt.savefig("PET_weighted_near_miss.svg", bbox_inches = 'tight')
plt.savefig("PET_weighted_near_miss.pdf", bbox_inches = 'tight')
plt.savefig("PET_weighted_near_miss.png", bbox_inches = 'tight', dpi=400)
# plt.xlim(0,10)
```
