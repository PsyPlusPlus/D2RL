## Load D2RL Raw Data

```python
root_folder = "../raw_data/D2RL/"
experiment_folder_list = ["Experiment-2lane_400m_D2RL_2022-09-10"]
```

```python
import os
import numpy as np
files = []

weighted_result_list = []
from tqdm import tqdm
for experiment_folder in experiment_folder_list:
    files = os.listdir(root_folder + experiment_folder)
    for file in tqdm(files):
        if file.endswith("npy"):
            try:
                weight_list_tmp = list(np.load(root_folder+experiment_folder+"/"+file).flatten())
                weighted_result_list.extend(weight_list_tmp)
            except:
                continue
```

## Save D2RL processed Data

```python
result_data_path = "./D2RL/weight_result.npy"
weighted_result_list = np.array(weighted_result_list)
np.save(result_data_path, weighted_result_list)
```

## Load NDE Raw Data

```python
root_folder = "../raw_data/NDE/"
experiment_folder_list = ["Experiment-2lane_400m_NDE_v1_2022-09-09", "Experiment-2lane_400m_NDE_v2_2022-09-09", "Experiment-2lane_400m_NDE_v3_2022-09-09"]
```

```python
files = []
import glob
for experiment_folder in experiment_folder_list:
    files_tmp = glob.glob(root_folder + experiment_folder + "/*.npy")
    files.extend(files_tmp)
result_path = root_folder + experiment_folder + "/result_fig/"
result_text_path = result_path + "result.txt"
result_data_path = result_path + "weight.npy"
result_npz_path = result_path + "weight.npz"
import os
os.makedirs(result_path, exist_ok=True)
weighted_result_list = []
total_length = 0
from tqdm import tqdm
crash_number = 0
tested_num = 0
expected_tested_num = 0



for file in tqdm(files):
    if file.endswith("npy"):
        try:
            result_tmp = np.load(file).flatten()
            crash_number += result_tmp[0]
            tested_num += result_tmp[1]
            expected_tested_num += result_tmp[2]
        except:
            continue

# print("Crash number:", crash_number)
# print("Total Number:", tested_num) 
# print("Expected Number:", expected_tested_num)
# print("NDE crash rate:", crash_number/tested_num)
```

## Save NDE processed Data
