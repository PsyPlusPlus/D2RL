## D2RL Near Miss Raw Data Process

```python
root_folder = "../raw_data/D2RL/"
path_list = ["Experiment-2lane_400m_D2RL_2022-09-10"]
```

```python
import glob
import json
from tqdm import tqdm
import random
from multiprocessing import Pool
def get_distance_ttc_from_one_json(json_path):
    try:
        with open(json_path, "r") as json_file:
            json_dict = json.load(json_file)
        weight = json_dict["weight_episode"]
        ttc_dict = json_dict["ttc_step_info"]
        distance_dict = json_dict["distance_step_info"]
        ttc_list = [ttc_dict[key] for key in ttc_dict if ttc_dict[key] < 10]
        distance_list = [distance_dict[key] for key in distance_dict if distance_dict[key] < 200]
    #     print(distance_list, ttc_list)
        return distance_list, ttc_list, weight
    except:
        return [], [], 0

def get_distance_ttc_from_experiments_D2RL(experiment_path):
    crash_folder_path = root_folder + experiment_path + "/crash/"
    safe_folder_path = root_folder + experiment_path + "/tested_and_safe/"
    crash_json_list = glob.glob(crash_folder_path + "*.json")
    safe_json_list = glob.glob(safe_folder_path + "*.json")
    distance_list_total, ttc_list_total, distance_weight_list, ttc_weight_list = [], [], [], []
    pool = Pool(10)
    total_near_miss_json_list = crash_json_list+safe_json_list
    for res in tqdm(pool.imap_unordered(get_distance_ttc_from_one_json, total_near_miss_json_list)):
        distance_list, ttc_list, weight = res
        distance_list_total.extend(distance_list)
        distance_weight_list.extend([weight]*len(distance_list))
        ttc_list_total.extend(ttc_list)
        ttc_weight_list.extend([weight]*len(ttc_list))
    return distance_list_total, ttc_list_total, distance_weight_list, ttc_weight_list
        
def get_distance_ttc_from_experiments_NDE(experiment_path):
    crash_folder_path = root_folder + experiment_path + "/crash/"
    safe_folder_path = root_folder + experiment_path + "/tested_and_safe/"
    crash_json_list = glob.glob(crash_folder_path + "*.json")
    safe_json_list = glob.glob(safe_folder_path + "*.json")
    distance_list_total, ttc_list_total, distance_weight_list, ttc_weight_list = [], [], [], []
    pool = Pool(10)
    total_near_miss_json_list = crash_json_list+safe_json_list
    for res in tqdm(pool.imap_unordered(get_distance_ttc_from_one_json, total_near_miss_json_list)):
        distance_list, ttc_list, weight = res
        distance_list_total.extend(distance_list)
        distance_weight_list.extend([weight]*len(distance_list))
        ttc_list_total.extend(ttc_list)
        ttc_weight_list.extend([weight]*len(ttc_list))
    return distance_list_total, ttc_list_total, distance_weight_list, ttc_weight_list
    
```

```python
distance_list_all_experiments, ttc_list_all_experiments = [], []
distance_weight_list_all_experiments, ttc_weight_list_all_experiments = [], []
for experiment_path in path_list:
    distance_list, ttc_list, distance_weight_list, ttc_weight_list = get_distance_ttc_from_experiments_D2RL(experiment_path)
    distance_list_all_experiments.extend(distance_list)
    ttc_list_all_experiments.extend(ttc_list)
    distance_weight_list_all_experiments.extend(distance_weight_list)
    ttc_weight_list_all_experiments.extend(ttc_weight_list)
    
```

```python
import numpy as np
save_folder = "./D2RL/"
np.save(save_folder + "NADE_distance_near_miss.npy", distance_list_all_experiments)
np.save(save_folder + "NADE_ttc_near_miss.npy", ttc_list_all_experiments)
np.save(save_folder + "NADE_distance_near_miss_weight.npy", distance_weight_list_all_experiments)
np.save(save_folder + "NADE_ttc_near_miss_weight.npy", ttc_weight_list_all_experiments)
```
